1.⁠ ⁠Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

```python
import numpy as np

# Define two custom numpy arrays A and B
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Stack A and B vertically
vertical_stack = np.vstack((A, B))

# Stack A and B horizontally
horizontal_stack = np.hstack((A, B))

# Display results
print("Array A:\n", A)
print("Array B:\n", B)
print("Vertical Stack:\n", vertical_stack)
print("Horizontal Stack:\n", horizontal_stack)
```

2.⁠ ⁠Find common elements between A and B. [Hint : Intersection of two sets]

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[3, 4], [5, 6]])

common_elements = np.intersect1d(A, B)
print("Common elements between A and B:", common_elements)
```

3.⁠ ⁠Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]

```python
# Define a custom numpy array A
A = np.array([[2, 5, 8], [10, 3, 7]])

# Define range
lower_bound = 5
upper_bound = 10

# Using boolean masking
filtered_values = A[(A >= lower_bound) & (A <= upper_bound)]

# Display result
print("Numbers in A between 5 and 10:", filtered_values)
```

4.⁠ ⁠Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

3.⁠ ⁠How to get the rows of a dataframe with row sum > 100?



```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df.head()
```

```python
# Filter rows where row sum > 100
filtered_df = df[df.sum(axis=1) > 100]

# Display the result
print(filtered_df)
```
